# TF-IDF Baseline

```python
dev_df = pd.DataFrame(dev_set)
```

```python
all_sentences = pd.concat([dev_df['sentence1'], dev_df['sentence2']], ignore_index=True).drop_duplicates()
```

```python
all_sentences
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from scipy.stats import pearsonr
```

```python
vectorizer = TfidfVectorizer(
    lowercase=True,
    stop_words='english',
    max_features=2000
)
vectorizer.fit(all_sentences)
```

```python
v1 = vectorizer.transform([dev_df.iloc[0]['sentence1']])
v2 = vectorizer.transform([dev_df.iloc[0]['sentence2']])
```

```python
cosine_similarity(v1,v2)
```

```python
def row_tfidf_similarity(row):
    v1 = vectorizer.transform([row['sentence1']])
    v2 = vectorizer.transform([row['sentence2']])
    return 5 * cosine_similarity(v1, v2)[0][0]
```

```python
dev_df['tf_idf_sim'] = dev_df.apply(row_tfidf_similarity, axis=1)
```

```python
dev_df
```

```python
pearsonr(dev_df['score'], dev_df['tf_idf_sim'])
```
